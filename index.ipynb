## Compare subway traffic by day of the week. Display this as a graph.

```python
#Convert to date
df.Date = pd.to_datetime(df.Date)
df['Weekday'] = df.Date.dt.dayofweek
df

#Get days of week frequencies
weekday_counts = df.groupby('Weekday').count()
weekday_counts = weekday_counts.loc[:, 'Linename']

#To Plot
y = list(weekday_counts)
x = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']

plt.figure(figsize=(10,6))
plt.bar(x, y, label='traffic')
plt.title('Subway Traffic by Day of Week')
plt.show()
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
y

#Find daily average of weekend traffic
sunday_val = y[0]
saturday_val = y[-1]
weekend_average = (sunday_val + saturday_val)/2

#Find daily average of weekday traffic
weekdays = y[1:-2]
weekdays_average = sum(weekdays)/len(weekdays)
weekdays_average

#
print('average weekday traffic: ' + str(weekdays_average))
print('average weekend traffic: ' + str(weekend_average))
print('more average daily traffic on weekdays')
```
